```python
income_df['Low Average Annual Pay (Less than $45000)'] = 0
income_df['Mid Average Annual Pay (Between $45000 & $70000)'] = 0
income_df['High Average Annual Pay (More than $70000)'] = 0
```

```python
income_df.loc[income_df['Average Annual Pay'].between(0,45000,'both'),'Low Average Annual Pay (Less than $45000)'] = 1
income_df.loc[income_df['Average Annual Pay'].between(45000,70000,'right'),'Mid Average Annual Pay (Between $45000 & $70000)'] = 1
income_df.loc[income_df['Average Annual Pay'].between(70000,4500000,'right'),'High Average Annual Pay (More than $70000)'] = 1
```

```python
income_df['Low Median Household Income (Less than $45000)'] = 0
income_df['Mid Median Household Income (Between $45000 & $70000)'] = 0
income_df['High Median Household Income (More than $70000)'] = 0
```

```python
income_df.loc[income_df['Median Household Income'].between(0,45000,'both'),'Low Median Household Income (Less than $45000)'] = 1
income_df.loc[income_df['Median Household Income'].between(45000,70000,'right'),'Mid Median Household Income (Between $45000 & $70000)'] = 1
income_df.loc[income_df['Median Household Income'].between(70000,4500000,'right'),'High Median Household Income (More than $70000)'] = 1
```

```python
income_df = income_df.merge(fips_df, how='inner',on='County')
```
