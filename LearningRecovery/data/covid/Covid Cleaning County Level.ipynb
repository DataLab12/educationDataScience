## Covid- County Level EDA


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_county_pop = pd.read_csv('covid_county_population_usafacts.csv')
df_county_pop.drop(columns = ['State'], inplace = True)
```

```python
df_county_pop = df_county_pop.rename(columns ={"countyFIPS" : "County #"})
df_county_pop
```

```python
df_county_deaths = pd.read_csv('covid_deaths_usafacts.csv')
df_county_deaths
```

```python
df_county_deaths = df_county_deaths[['countyFIPS','County Name','2020-09-28','2020-10-30','2021-01-29','2021-07-28','2022-01-29','2022-07-29','2022-10-30']]
df_county_deaths
```

```python
df_county_deaths = df_county_deaths.rename(columns ={"2020-09-28": "Confirmed Deaths 09/28/20",
                                                     "2020-10-30" : "Confirmed Deaths 10/30/20",
                                                     "2021-01-29" : "Confirmed Deaths 01/29/21",
                                                     '2021-07-28':'Confirmed Deaths 07/28/2021',
                                                     '2022-01-29':'Confirmed Deaths 01/29/2022',
                                                     '2022-07-29':'Confirmed Deaths 07/29/2022',
                                                     '2022-10-30':'Confirmed Deaths 10/30/2022',
                                                    "countyFIPS" : "County #"})
```

```python
df_county_deaths.info()
```

```python
df_county_con = pd.read_csv('covid_confirmed_usafacts.csv')
df_county_con
```

```python
df_county_con = df_county_con[['countyFIPS', 'County Name','2020-09-28','2020-10-30','2021-01-29','2021-07-28','2022-01-29','2022-07-29','2022-10-30']]
df_county_con
```

```python
df_county_con = df_county_con.rename(columns ={"2020-09-28": "Confirmed Cases 09/28/20",
                                               "2020-10-30" : "Confirmed Cases 10/30/20",
                                               "2021-01-29" : "Confirmed Cases 01/29/21",
                                               '2021-07-28':'Confirmed Cases 07/28/2021',
                                               '2022-01-29':'Confirmed Cases 01/29/2022',
                                               '2022-07-29':'Confirmed Cases 07/29/2022',
                                               '2022-10-30':'Confirmed Cases 10/30/2022',
                                              "countyFIPS" : "County #"})
```

```python
df_county_con
```

```python
df_merge = pd.merge(df_county_deaths, df_county_con, how = "right", on= "County #")
df_merge = pd.merge(df_merge, df_county_pop, how = "left", on = "County #")
```

```python
df_merge.drop(columns = ['County Name_x', 'County Name_y', 'County Name'], inplace = True)
df_merge.rename(columns = {"population": "County Population"}, inplace = True)
df_merge.drop(index=0)
```

```python
#df_merge['County #'] = df_merge['County #'].astype(float)
df_merge['Confirmed Deaths 09/28/20'] = df_merge['Confirmed Deaths 09/28/20'].astype(float)
df_merge['Confirmed Deaths 10/30/20'] = df_merge['Confirmed Deaths 10/30/20'].astype(float)
df_merge['Confirmed Deaths 01/29/21'] = df_merge['Confirmed Deaths 01/29/21'].astype(float)
df_merge['Confirmed Cases 09/28/20'] = df_merge['Confirmed Cases 09/28/20'].astype(float)
df_merge['Confirmed Cases 10/30/20'] = df_merge['Confirmed Cases 10/30/20'].astype(float)
df_merge['Confirmed Cases 01/29/21'] = df_merge['Confirmed Cases 01/29/21'].astype(float)
df_merge['County Population'] = df_merge['County Population'].astype(float)
```

```python
columns = ['County', 'County Name', 'County #', 'Public Health Region', 'Health Service Region']

cc = pd.read_csv('county_conversion.csv', names = columns, header = 0)

cc.drop(columns = ['Public Health Region', 'Health Service Region', 'County Name','County'], inplace = True)

cc
```

```python
df_merge = pd.merge(df_merge, cc, how = "right", on = "County #")
```
