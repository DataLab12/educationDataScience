***Data Reading:***

```python
# read in the csv file
df_d = pd.read_excel('ada_district.xlsx')

print(df_d.info())
df_d.head()
```

```python
cols_old = df_d.columns.to_list()
cols_new = ['County #',
'County Name',
'District #',
'District Name',
'2004-2005 ADA',
'2005-2006 ADA',
'2006-2007 ADA',
'2007-2008 ADA',
'2008-2009 ADA',
'2009-2010 ADA',
'2010-2011 ADA',
'2011-2012 ADA',
'2012-2013 ADA',
'2013-2014 ADA',
'2014-2015 ADA',
'2015-2016 ADA',
'2016-2017 ADA',
'2017-2018 ADA',
'2018-2019 ADA',
'2019-2020 ADA',
'2020-2021 ADA',
'2021-2022 ADA']

df_d.rename(columns=dict(zip(cols_old, cols_new)), inplace=True)
```

```python
df_d.replace(',', '', regex=True, inplace=True)
df_d[cols_new[4:]] = df_d[cols_new[4:]].astype('float')
```

```python
cols_keep=['County #',
'District #',
'2018-2019 ADA',
'2019-2020 ADA',
'2020-2021 ADA',
'2021-2022 ADA']

df_d[cols_keep].info()
```

```python
hold=pd.DataFrame()
hold['County Name'] = df_d['County Name']
```

```python
hold.drop_duplicates(inplace=True)
```

```python
df_d.drop(['County Name','District #','District Name'],axis=1, inplace=True)
```

```python
df_c=df_d.groupby('County #')[['2017-2018 ADA','2018-2019 ADA','2019-2020 ADA','2020-2021 ADA','2021-2022 ADA']].sum()
df_c
```

```python
df_c.reset_index(inplace=True)
df_c
```

```python
df_c.rename(columns={'County #':'County'},inplace=True)
```

```python
columns = ['County', 'County Name', 'County #', 'Public Health Region', 'Health Service Region']

cc = pd.read_csv('county_conversion.csv', names = columns, header = 0)

cc.drop(columns = ['Public Health Region', 'Health Service Region', 'County Name'], inplace = True)

cc
```

```python
df_c = df_c.merge(cc, how = "inner", on = "County")
```
